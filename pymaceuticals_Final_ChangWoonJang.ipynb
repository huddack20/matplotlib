```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

tv_mean = cleaned_clinical_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
tv_median = cleaned_clinical_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
tv_variance = cleaned_clinical_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
tv_std = cleaned_clinical_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()

# SEM is the standard deviation of mean of random samples drawn from the original population. 
# Just as the sample SD (s) is an estimate of variability of observations, SEM is an estimate 
# of variability of possible values of means of samples.
tv_sem = cleaned_clinical_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

stat_clinical_df = pd.concat([tv_mean, tv_median, tv_variance, tv_std, tv_sem], 
                              axis=1, 
                              keys=["Tumor Volume Mean (mm3)", "Tumor Volume Median (mm3)",
                              "Tumor Volume Variance (mm3)", "Tumor Volume Std (mm3)", 
                              "Tumor Volume SEM (mm3)"])

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

stat_clinical_df
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
final_tumor_vol_df = cleaned_clinical_df.sort_values(by=['Mouse ID', 'Timepoint'])

# Capomulin, Ramicane, Infubinol, and Ceftamin
Regimem_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Start by getting the last (greatest) timepoint for each mouse
final_tumor_vol_max_df = final_tumor_vol_df.groupby(['Mouse ID', 'Drug Regimen'])['Timepoint'].max()
final_tumor_vol_max_df = final_tumor_vol_max_df.reset_index()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
final_tumor_vol_max_df = final_tumor_vol_max_df.rename(columns = {'Timepoint': 'Max_Timepoint'})
merged_df = pd.merge(cleaned_clinical_df, final_tumor_vol_max_df, on='Mouse ID')
merged_df = merged_df.rename(columns = {'Drug Regimen_x': 'Drug Regimen'})

merged_df.head()
```

```python

# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_stat = {"Treatment": [],
              "Lower Bound": [],
              "Upper Bound": [],
              "IQR": []}

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for treatment in treatments:
    
    temp_df = merged_df.loc[merged_df['Timepoint'] == merged_df['Max_Timepoint']]
    
    Q1 = np.quantile((temp_df.loc[temp_df['Drug Regimen'] == treatment])['Tumor Volume (mm3)'], 0.25)
    Q3 = np.quantile((temp_df.loc[temp_df['Drug Regimen'] == treatment])['Tumor Volume (mm3)'], 0.75)

    IQR = Q3 - Q1
    
    tumor_stat["Treatment"].append(treatment)
    
    Upper = Q3+1.5*IQR
    tumor_stat["Upper Bound"].append(round(Upper,2))
    
    Lower = Q1-1.5*IQR
    tumor_stat["Lower Bound"].append(round(Lower,2))
    
    tumor_stat["IQR"].append(round(IQR, 2))

  # Above Upper bound
  # Below Lower bound
print("***Summary of Statistics***")
print(f"Treatment: {tumor_stat['Treatment']}")
print(f"Upper Bound: {tumor_stat['Upper Bound']}")
print(f"Lower Bound: {tumor_stat['Lower Bound']}")
print(f"IQR: {tumor_stat['IQR']}")
print("\n")
        
    # Locate the rows which contain mice on each drug and get the tumor volumes
    
    
    # add subset 
    
    
    # Determine outliers using upper and lower bounds
lower_outlier = {"Treatment": [], 
                "# of Outliers Below Lower Bound": []}
upper_outlier = {"Treatment": [], 
                "# of Outliers Above Upper Bound": []}

for i, treatment in enumerate(treatments):
    condition = (merged_df.loc[merged_df['Drug Regimen'] == treatment])['Tumor Volume (mm3)']
    
    # outlier below lower bound
    lower = (condition < float(tumor_stat["Lower Bound"][i])).value_counts()
    # outlier above upper bound
    upper = (condition > float(tumor_stat["Upper Bound"][i])).value_counts()

    if len(lower) == 2 or len(upper) == 2:

        if len(lower) == 2:
            lower_outlier["Treatment"].append(treatment)
            lower_outlier["# of Outliers Below Lower Bound"].append(round(lower[1], 2))
            
        if len(upper) == 2:
            upper_outlier["Treatment"].append(treatment)
            upper_outlier["# of Outliers Above Upper Bound"].append(round(upper[1]))

if (len(lower_outlier["# of Outliers Below Lower Bound"])) != 0:
    print(lower_outlier)
if (len(upper_outlier["# of Outliers Above Upper Bound"])) != 0:
    print(upper_outlier)
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

Capomulin_df = cleaned_clinical_df.loc[cleaned_clinical_df["Drug Regimen"] == 'Capomulin']

# Mean value of Tumor Volume grouped by mouse weight for Capomulin Regimen
TumorVol_Weight_Capomulin_mean = Capomulin_df.loc[Capomulin_df['Drug Regimen'] == 'Capomulin'].groupby('Weight (g)')['Tumor Volume (mm3)'].mean()

# Reset Index
TumorVol_Weight_Capomulin = TumorVol_Weight_Capomulin_mean.reset_index()

plt.figure(figsize=(5, 5))
plt.scatter(TumorVol_Weight_Capomulin['Weight (g)'], TumorVol_Weight_Capomulin['Tumor Volume (mm3)'], s=100, color='black')

plt.title('Capomulin', fontsize=20)
plt.xlabel('Weight (g)', fontsize=15)
plt.xticks(fontsize=15)
plt.ylabel('Average Tumor Volume (mm3)', fontsize=15)
plt.yticks(fontsize=15)

plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

x = TumorVol_Weight_Capomulin['Weight (g)']
y = TumorVol_Weight_Capomulin['Tumor Volume (mm3)']

# Calculate the correlation coefficient between mouse weight and average tumor volume
# Sicpy Stat
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x, y)
print(f'*Correlation values r, p-value by "Scipy Stat": {rvalue}, {pvalue}')

corr_df = pd.DataFrame({"Weight (g)": x,
                      "Average Tumor Volume (mm3)": y})

# Pandas corr function
correlation_matrix_df = corr_df.corr(method="pearson")
print('\n--Correlation Matrix by "Pandas corr function"')
print(f'{correlation_matrix_df}\n')
print(f'*Correlation r value by "Pandas corr function": {correlation_matrix_df.iloc[0, 1]}')

# Linear Regression Formula
y_pred = [] # List to store prediction values
y_pred = round(slope, 2)*x + round(intercept, 2)

plt.figure(figsize=(5, 5))
plt.scatter(TumorVol_Weight_Capomulin['Weight (g)'], TumorVol_Weight_Capomulin['Tumor Volume (mm3)'], s=100, color='black')
plt.plot(x, y_pred, color='r', linestyle='--')
plt.xlabel("Weight (g)", fontsize=15)
plt.ylabel("Average Tumor Volume (mm3)", fontsize=15)
plt.title("Capomulin", size=20)
plt.annotate(f"y = {round(slope, 2)}x + {round(intercept, 2)}", (15.2,43), fontsize=15, color="red")
plt.annotate(" -Regression line-", (15,44), fontsize=15, color="red")

plt.show()
```
